```python
import os
from llama_index.core.llms import ChatMessage, MessageRole
from llama_index.llms.groq import Groq
from dotenv import load_dotenv
load_dotenv()
```

```python
api_key = os.getenv("GROQ_API_KEY")
```

```python
llm = Groq(
    model="llama-3.3-70b-versatile",
    temperature=0,
)
```

```python
def chat():
    history = [
        ChatMessage(role=MessageRole.SYSTEM, content="You are a helpful chatbot, be concise and accurate")
    ]

    print("Llamaindex Chatbot. type 'exit' to quit \n")

    while True:
        user_input = input("You: ").strip()
        if user_input.lower() == "exit":
            break
        if not user_input:
            continue
        # add user message
        history.append(ChatMessage(role=MessageRole.USER, content=user_input))

        #call LLM with chat history
        response = llm.chat(messages=history)
        answer = response.message.content

        print(f"Bot: {answer}\n")

        history.append(ChatMessage(role=MessageRole.ASSISTANT, content=answer))
```

```python
chat()
```
